## 1.Time Series Analysis
#### 1.1 important concepts in time series analysis that describe the behavior of the data over time are:

     * Stationarity and
     * non-stationarity 
 
#### 1.2 Stationary Time Series:
* A stationary time series is one in which the statistical properties, such as the mean, variance, and autocorrelation, remain constant over time. In other words, the properties of the time series do not depend on the specific time at which they are observed.

* Stationary time series have the following characteristics:

* Constant mean: The mean of the time series remains constant over time.
* Constant variance: The variance of the time series remains constant over time.
* Constant autocovariance: The autocovariance between any two observations in the series depends only on the time lag between them and not on the specific time at which they are observed.
* Stationary time series are relatively easier to analyze and model because their behavior is more predictable. Many statistical models and techniques, such as autoregressive integrated moving average (ARIMA), assume stationarity in the data.

#### 1.3 Non-Stationary Time Series:

* A non-stationary time series is one in which the statistical properties change over time. This means that the mean, variance, or autocorrelation may exhibit trends, cycles, or other patterns.

* Non-stationary time series can have various characteristics:

* Trend: The series shows a long-term increase or decrease over time.
* Seasonality: The series exhibits regular patterns or fluctuations that repeat over fixed time intervals.
* Changing variance: The variance of the series changes over time.
* Unit root: A unit root is present if the series has a stochastic trend, meaning it does not revert to a constant level over time.
* Non-stationary time series pose challenges in analysis and modeling because their properties change, making it difficult to apply traditional statistical techniques. However, non-stationary data can often be transformed into stationary data through techniques like differencing, logarithmic transformation, or seasonal adjustment.

* To analyze non-stationary time series, specialized models like autoregressive integrated moving average (ARIMA) with differencing (the "I" component) or more advanced models like autoregressive integrated moving average with exogenous variables (ARIMAX) or vector autoregression (VAR) can be used.

### 1.4 Components Of Time Series And It's Explanation

* As we define above Time series data refers to a sequence of observations or measurements that are recorded at regular intervals over time. Time series analysis is a statistical method used to analyze such data and uncover patterns, trends, and dependencies.

#### There are different types of time series based on the characteristics of the data and the patterns it exhibits. Here are some common types of time series:

#### Trend:
* A time series with a trend shows a long-term increase or decrease in values over time. The trend can be linear or nonlinear. For example, a time series of monthly sales data might show an increasing trend if sales are consistently growing over time.

#### Seasonal:
* A seasonal time series exhibits regular patterns or fluctuations that repeat over fixed time intervals. These patterns can be daily, weekly, monthly, or yearly. For instance, a time series of monthly ice cream sales data might show a seasonal pattern with higher sales during the summer months and lower sales during the winter months.

#### Cyclical:
* A cyclical time series displays patterns that are not as regular as seasonal patterns and have longer durations. These cycles do not have fixed time intervals and can span several years. Economic indicators such as stock market prices or business cycles often exhibit cyclical patterns.

#### Irregular or Residual:
* An irregular time series, also known as a residual or random component, represents the unpredictable or random fluctuations that cannot be explained by the trend, seasonal, or cyclical patterns. It includes random noise, outliers, and unpredictable events. This component is often modeled as random variations around the other components.

#### Autoregressive Integrated Moving Average (ARIMA): 
* ARIMA is a popular time series model that combines autoregressive (AR), differencing (I), and moving average (MA) components. It is suitable for analyzing time series data that exhibit both trend and seasonality. ARIMA models can capture dependencies between current and past observations.

* Long Short-Term Memory (LSTM): LSTM is a type of recurrent neural network (RNN) that is commonly used for time series analysis. It is effective at capturing long-term dependencies and is particularly useful when dealing with sequential data. LSTM models have been successfully applied to various time series tasks, including forecasting and anomaly detection.

# Lets perform Dichey Fuller Test

````python
from statsmodels.tsa.stattools import adfuller
print("Results of Dickey-fuller Test")

dftest=adfuller(df['Sum of Total Malaria Case'],autolag="AIC")

dfoutput=pd.Series(dftest[0:4],index=["Test Statistic","P-value","lags Used","Number of Observation Used"])

                                   
for key,value in dftest[4].items():
    dfoutput['Critical value (%s) '%key]=value
print(dfoutput)
                                      
                                
````

The results of the Dickey-Fuller test provide information about the stationarity of a time series. Stationarity is an important concept in time series analysis, and it refers to the property where the statistical properties of a time series, such as mean and variance, remain constant over time.

Here's the interpretation of the results:

Test Statistic: The test statistic value is 0.593373. This value is used to assess the stationarity of the time series. In order to determine whether the time series is stationary, the test statistic is compared to critical values.

P-value: The p-value associated with the test statistic is 0.987451. The p-value represents the probability of obtaining the test statistic value or a more extreme value, assuming the null hypothesis is true. In the Dickey-Fuller test, the null hypothesis is that the time series is non-stationary. A p-value greater than a chosen significance level (e.g., 0.05) suggests that there is strong evidence to not reject the null hypothesis of non-stationarity.

Lags Used: The number of lags used in the test is 7. Lags refer to the number of past observations included in the test. In this case, 7 lags were used to account for potential autocorrelation in the time series.

Number of Observation Used: The number of observations used in the test is 112. This indicates the number of non-missing values in the time series that were used for the analysis.

Critical Values: The critical values represent thresholds that are used to determine whether the test statistic is significantly different from the expected value under the null hypothesis. In this case, three critical values are provided: 1%, 5%, and 10%. If the test statistic is lower than these critical values, it provides evidence to reject the null hypothesis of non-stationarity.

Based on the results provided, the test statistic of 0.593373 is higher than all the critical values. Additionally, the p-value of 0.987451 is much greater than the typical significance level of 0.05. These results suggest that there is not enough evidence to reject the null hypothesis of non-stationarity. In other words, the time series is likely non-stationary.

It's important to note that the interpretation of the Dickey-Fuller test results may vary depending on the specific context and requirements of the analysis. Additionally, further analysis and consideration of other factors, such as data trends and seasonality, may be required to fully understand the behavior of the time series.

# a. Trend Analysis:
* Non-stationarity often indicates the presence of a trend in the data. You can perform trend analysis to identify and analyze any underlying long-term patterns or trends in the time series. This can involve techniques such as linear regression, moving averages, or decomposition methods like seasonal decomposition of time series (STL decomposition).

# b. Seasonality Analysis: 
* If the time series exhibits recurring patterns or seasonality, you can explore techniques to identify and analyze these patterns. This can involve methods such as autocorrelation analysis, Fourier analysis, or seasonal decomposition techniques.

# c. Differencing: 
* Differencing is a common technique used to transform a non-stationary time series into a stationary one. By taking differences between consecutive observations, you can eliminate or reduce trends and achieve stationarity. Differencing can be performed once or multiple times until the resulting series is stationary.

# d. Autocorrelation Analysis:
* Autocorrelation analysis helps to identify the presence of serial correlation or dependencies between observations in a time series. # e. Seasonal Adjustment: 
* If the time series exhibits seasonal patterns, you can apply seasonal adjustment techniques to remove the seasonal component and focus on the underlying trend and irregular components. Methods such as seasonal differencing or seasonal decomposition (e.g., X-12-ARIMA) can be employed for this purpose.

# f. Time Series Modeling:
* Non-stationary time series can still be modeled using various techniques such as autoregressive integrated moving average (ARIMA) models or their extensions (e.g., SARIMA for seasonal data). These models are designed to handle non-stationary data by incorporating differencing and lagged dependencies.

# a. Trend analysis
* is a technique used to identify and analyze underlying long-term patterns or trends in a time series. When dealing with a non-stationary time series, where the statistical properties change over time, identifying and understanding the trend becomes particularly important. Here are some methods commonly used for trend analysis:

# 1.Moving Averages: 
* Moving averages are widely used to smooth out fluctuations in a time series and reveal the underlying trend. A moving average is calculated by taking the average of a fixed window of observations. By adjusting the window size, you can control the level of smoothing. Moving averages can help in visualizing and understanding the overall trend of the data.

# 2.Regression Analysis:
* Regression analysis can be used to model the relationship between the time variable (e.g., month or year) and the variable of interest in the time series. This allows you to estimate the trend and understand how it changes over time. Simple linear regression or more complex regression models, such as polynomial regression or exponential regression, can be applied depending on the nature of the trend.

# 3.Decomposition: 
* Time series decomposition separates a time series into its constituent components: trend, seasonality, and residual (or irregular) component. Decomposition can be performed using techniques such as seasonal decomposition of time series (STL decomposition), classical decomposition, or X-12-ARIMA. By isolating the trend component, you can analyze its behavior and understand the underlying long-term patterns.

# 4.Smoothing Techniques:
* Smoothing techniques, such as exponential smoothing or the Hodrick-Prescott filter, can be used to estimate the trend component of a time series. These techniques utilize weighted averages or mathematical algorithms to extract the trend from the data. Exponential smoothing assigns weights that decrease exponentially as observations become older, while the Hodrick-Prescott filter separates the trend and cyclical components based on a specific mathematical criterion.

# 5.Non-linear Models:
* Non-linear models can capture complex and non-linear trends in a time series. Techniques such as spline regression, loess regression, or generalized additive models (GAMs) can be employed to model and analyze non-linear trends. These models can be more flexible in capturing various patterns that may be missed by simpler linear models.

````python
# Perform time series decomposition
decomposition = sm.tsa.seasonal_decompose(total_case["Sum of Total Malaria Case"], model='additive')

# Access the decomposition components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Print the decomposition components
print("=========================================================================")
print("Trend component:")
print(trend)
print("=========================================================================")
print("Seasonal component:")
print(seasonal)
print("=========================================================================")
print("Residual component:")
print(residual)
````

````python
Decoposed_data=pd.concat([total_case["Sum of Total Malaria Case"],trend,seasonal,residual],axis=1)
Decoposed_data.columns=['total_malaria_case','Trend','Seasonal','Residual']
#Decoposed_data.to_csv(Decoposed_data,sep=',',index=False,encoding='utf-8')
Decoposed_data.head(50)
````

The model parameter in the seasonal_decompose function determines the type of decomposition model used to separate the time series into its components: trend, seasonal, and residual.

There are two options for the model parameter:

'additive': In the additive model, it is assumed that the seasonal component has an additive relationship with the trend and residual components. This means that the seasonal component is added to the trend and residual components to reconstruct the original time series. Mathematically, the decomposition can be represented as:

Copy
y(t) = Trend(t) + Seasonal(t) + Residual(t)
```

In this model, the seasonal component represents the average deviation from the trend at each point in the time series.

'multiplicative': In the multiplicative model, it is assumed that the seasonal component has a multiplicative relationship with the trend and residual components. This means that the seasonal component is multiplied by the trend and residual components to reconstruct the original time series. Mathematically, the decomposition can be represented as:

Copy
y(t) = Trend(t) * Seasonal(t) * Residual(t)
```

In this model, the seasonal component represents the proportional deviation from the trend at each point in the time series.
The choice between the additive and multiplicative models depends on the underlying characteristics of the time series. Here are some guidelines:

Additive Model: The additive model is appropriate when the magnitude of the seasonal fluctuations in the time series does not depend on the level of the trend. In other words, the seasonal component contributes a constant amount of variation to the time series, regardless of whether the trend is high or low.

Multiplicative Model: The multiplicative model is suitable when the magnitude of the seasonal fluctuations is proportional to the level of the trend. In this case, the seasonal component contributes a varying amount of variation to the time series, depending on the level of the trend.

It's important to note that the choice of the model parameter may require some experimentation and domain knowledge. You can try both models and assess which one provides a better representation of the underlying patterns in your specific time series data.

A spike at lag 0 may indicate the presence of a trend or other non-stationary patterns but, in our case it is not lag 0.
